```python
genre_year_stats = df.groupby(["Genre","ReleaseYear"]).agg({
  "Global_BoxOfficeUSD" : "mean",
  "BudgetUSD" : "mean",
  "IMDbRating" : "mean",
  "RottenTomatoesScore" : "mean" # aggregation functions used to calculate mean(average).
}).reset_index() # Converts grouped index back into normal columns.
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data= genre_year_stats, x = "ReleaseYear", y = "Global_BoxOfficeUSD", hue="Genre")
plt.title("Average Global Box Office By Genre over years")
plt.show()
# plt.figure : Creates a new plotting canvas with custom size. 
# Meaning : 15  width (in inches) 10  height (in inches)
# Without hue="Genre"  only one line (overall average)
# With hue  multi-line comparison across genres
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data= genre_year_stats.query("Genre == 'Action'"), x = "ReleaseYear", y = "Global_BoxOfficeUSD", hue="Genre")
plt.title("Average Global Box Office By Genre over years")
plt.show()

# .query : for any type of Genre as now we have take 'Action'.
```

```python
top_actors = df.groupby("LeadActor") ["IMDbRating"].mean().sort_values(ascending=False).head(10)
# Groups dataset by LeadActor. 
# Then select only IMDbRating column from each group. 
# Then Calculate average IMDbRating. 
```

```python
top_actors
```

```python
plt.figure(figsize=(15,8))
top_actors.plot(kind="bar")
plt.title("Top 10 Actors by AVG IMDb Rating")
plt.show
```

```python
import pandas as pd # For data handling and analysis
from tabulate import tabulate # For converting data into table format

df = pd.read_csv("movies_dataset.csv", on_bad_lines="skip") # read movies_dataset.csv into DataFrame(df) and skip the rows with corrupted or wrong column counts.

table = tabulate(df, headers="keys", tablefmt="grid", showindex=False) 
# df - Data to convert into table. 
# headers="keys" : Uses DataFrame column names as headers.
# tablefmt="grid" : Displays table in grid format. 
# showindex=False : Removes row index column from output

with open("data.txt", "w", encoding="utf-8") as f:
    f.write(table)
# "w"  write mode. 
# encoding="utf-8" : Supports special characters.
# with statement : Automatically closes file
# f.write(table) : Writes formatted table into the text file
```
